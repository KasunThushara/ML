### Data Preprosessing 

```python
import os
data_set_path=os.listdir('House_Room_Dataset')
print("There are subfolders", len(data_set_path))
print(data_set_path)
```

```python
rooms=[]
for item in data_set_path:
    all_types=os.listdir('House_Room_Dataset'+'/'+item)
    #print(item)
    for room in all_types:
        rooms.append((item,str('House_Room_Dataset'+'/'+item)+'/'+room))
print(rooms)
    
```

```python
import pandas as pd
rooms_df=pd.DataFrame(data=rooms,columns=['room type','image'])
print(rooms_df.head())
```

```python
print("Total",len(rooms_df))
rooms_count=rooms_df['room type'].value_counts()
print("rooms in each group: ")
print(rooms_count)
```

```python
import cv2
import numpy as np
path='House_Room_Dataset/'
im_size=300
images=[]
labels=[]
for i in data_set_path:
    data_path=path+str(i)
    filenames=[i for i in os.listdir(data_path)]
    for f in filenames:
        img=cv2.imread(data_path+'/'+f)
        img=cv2.resize(img,(im_size,im_size))
        images.append(img)
        labels.append(i)
images=np.array(images)
images=images.astype('float32')/255.0
images.shape
    #print(data_path)
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
y=rooms_df['room type'].values
y_labelencoder=LabelEncoder()
y=y_labelencoder.fit_transform(y)
print(y)
```

```python

one_hot_encoded_data = pd.get_dummies(rooms_df, columns = ['room type'])
print(one_hot_encoded_data)
```

```python
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split

images,one_hot_encoded_data=shuffle(images,one_hot_encoded_data,random_state=1)
train_x,test_x,train_y,test_y=train_test_split(images,one_hot_encoded_data,test_size=0.2,random_state=123)
print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)
```
